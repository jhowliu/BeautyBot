```python
import jieba
import json
import re
```

```python
with open("corpus.json") as fp:
    lines = fp.readlines()
    
with open("stop_words.txt") as fp:
    stop_words = fp.readlines()

lines = map(lambda x: x.replace("\n", ""), lines)
stop_words = map(lambda x: x.replace("\n", ""), stop_words)
stop_words = "|".join(stop_words)

raw_text = ""
for line in lines:
    raw_text += line
```

```python
corpus = json.loads(raw_text)
jieba.set_dictionary('dict.txt.big')
stop_words = stop_words.decode('utf-8')
```

```python
print corpus['articles'][0]['messages']
```

```python
is_threw_away = re.compile(ur"[神人|公告]")
picture_index = re.compile(r"imgur.com\/(\w+)")
punctuation_cleaner = re.compile(ur"[ =￣ω￣><=\w\:\s+\.\!\/_?～,$%^*(+\"\'\]\[——！，。？、~@#￥%……&*（）：；《）《》“”()»〔〕-]+")
stop_words_cleaner = re.compile(stop_words)
url = "http://imgur.com/"
```

```python
articles = corpus["articles"]
```

```python
raw_data = {}

for article in articles: 
    content = article["content"]
    title = punctuation_cleaner.sub('', article["article_title"])
    msgs = article['messages']
    
    if is_threw_away.search(title) is None:
        print title
        indices = picture_index.findall(content)
        raw_data[title] = {}
        raw_data[title]["comment"] = [stop_words_cleaner.sub("", punctuation_cleaner.sub("", msg['push_content'])) for msg in msgs]
        raw_data[title]["pictures"] = ["".join([url, index, ".jpg"]) for index in indices]
```
